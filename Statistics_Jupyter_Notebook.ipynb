#### Question 4

What is correlation? Give an example with a dataset & graphical representation on jupyter Notebook.
#### Answer:
Correlation is a statistical measure that expresses the extent to which two variables are linearly related. It is a value between -1 and 1, where:

1 indicates a perfect positive linear relationship,

-1 indicates a perfect negative linear relationship, and

0 indicates no linear relationship between the variables.

```python
# importing the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Creating a dataset
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)  # y is linearly related to x

# Creating a pandas DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Calculating the correlation
correlation_matrix = data.corr()

# Plotting the dataset and showing the correlation coefficient
plt.figure(figsize=(8, 5))
plt.scatter(data['X'], data['Y'], color='blue', label=f"Correlation coefficient: {correlation_matrix.iloc[0, 1]:.2f}")
plt.title("Scatter plot of X vs Y showing Correlation")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.legend()
plt.grid(True)
plt.show()

correlation_matrix
```
